```python
AI_compute.rename(columns = {'Relative # Job postings' : 'Relative Job postings'}, inplace=True)`
```

```python
AI_compute['Year'] = pd.to_datetime(AI_compute['Date']).dt.year
```

```python
grouped = AI_compute.groupby('Country')
```

```python
# Create the line chart
fig, ax = plt.subplots(figsize=(10, 6))

for country, group in grouped:
    group.plot(x="Date", y="Relative Job postings", ax=ax, label=country)

    
# Customize the plot
plt.xlabel("Date")
plt.ylabel("Job Posting")
plt.title("Comparison of Job Posting by Country")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend()

```
